# Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor #Importing VIF library

def VarianceInflationFactor(dataset): #Creating a function for VIF
    vif=pd.DataFrame() #Creating a dataframe called "vif"
    vif["Variables"]=dataset.columns #Assigning numerical column names to the varible named "Variables"
    vif["VIF"]=[variance_inflation_factor(dataset.values,i) for i in range(dataset.shape[1])] #Calculating VIF values for each numeric columns using one liner for loop
    return vif #Returning the dataframe
```

```python
VarianceInflationFactor(dataset[Quan])
```

# t-Test

#Independent t-Test: Unpaired t-Test (Different groups[Central vs others] but same condition[salary])

```python
from scipy.stats import ttest_ind
central=dataset[dataset["ssc_b"]=="Central"]["salary"]
others=dataset[dataset["ssc_b"]=="Others"]["salary"]
ttest_ind(central,others)
```

```python
#There is a large difference in the salary between students studied in Central board and others since the pvalue is 61%
```

#Dependent t-Test: Paired T-Test (Same group[Commerce] but different condition[etest_p vs mba_p])

```python
from scipy.stats import ttest_rel
etest=dataset[dataset["hsc_s"]=="Commerce"]["etest_p"]
mba=dataset[dataset["hsc_s"]=="Commerce"]["mba_p"]
ttest_rel(etest,mba)
```

```python
#The pvalue is very high (246%) hence there is large difference between etest_p and mba_p for students studying in Commerce group
```

# Analysis Of Variance

#One way classification

```python
import scipy.stats as stats
stats.f_oneway(dataset["ssc_p"],dataset["hsc_p"],dataset["degree_p"],dataset["etest_p"],dataset["mba_p"],dataset["salary"])
```

#Two way classification

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model=ols("dataset['salary']~C(dataset['degree_t'])+C(dataset['specialisation'])+C(dataset['degree_t']):C(dataset['specialisation'])",data=dataset).fit()
anova_lm(model,type=2)
```
